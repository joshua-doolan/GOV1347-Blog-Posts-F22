```python
col_names = pd.read_csv("delux_data/cand_link/ccl_header_file.csv").columns

cand_link_dfs = []
for i in range(18,22+1,2):
    cand_link_df = pd.read_csv(f"delux_data/cand_link/ccl{i}.txt", delimiter="|", names=col_names)
    cand_link_df["CYCLE"] = 2000 + i
    cand_link_dfs.append(cand_link_df)

cand_link_df = pd.concat(cand_link_dfs)

cand_link_df.head()
```

```python
col_names = pd.read_html("Current campaigns for House and Senate file description _ FEC.html")[0].iloc[1:,0]

cand_summ_dfs = []
for i in range(18,22+1,2):
    cand_summ_df = pd.read_csv(f"delux_data/cand_summ/webl{i}.txt", delimiter="|", names=col_names)
    cand_summ_df["CYCLE"] = 2000 + i
    cand_summ_dfs.append(cand_summ_df)

cand_summ_df = pd.concat(cand_summ_dfs)

cand_summ_df.head()
```

```python
party_table = pd.merge(cand_link_df, cand_summ_df, on=["CAND_ID", "CYCLE"], how="inner")[["CYCLE", "CMTE_ID", "CAND_PTY_AFFILIATION", "CAND_NAME", "CAND_OFFICE_DISTRICT"]]
```

```python
results_df = pd.read_csv("lite_data/incumb_dist_1948-2022 (2).csv")
results_df.head()
# results_df[(results_df["year"] == 2018) & (results_df["state"] == "Alaska")]
```

```python
competitive_races_df = pd.read_csv("expert_rating.csv")
```

```python
super_competitive_df = competitive_races_df[((competitive_races_df["avg_rating"] > 1.5) & (competitive_races_df["avg_rating"] < 5.5))]
super_competitive_df
```

```python
super_competitive_df.loc[super_competitive_df["district"] == "AL", "district"] = 0

super_competitive_df["district"] = super_competitive_df["district"].astype(int)
```

```python
cont_grouped = train_contributions.groupby(["cycle", "state_full", "CAND_OFFICE_DISTRICT", "RepParty"]).agg({"total": "sum", "count": "sum"}).reset_index()

# Create Columns for Republican Party total and count, and Democratic Party total and count

cont_grouped["RepTotal"] = cont_grouped["total"] * cont_grouped["RepParty"]
cont_grouped["RepCount"] = cont_grouped["count"] * cont_grouped["RepParty"]
cont_grouped["DemTotal"] = cont_grouped["total"] * (1 - cont_grouped["RepParty"])
cont_grouped["DemCount"] = cont_grouped["count"] * (1 - cont_grouped["RepParty"])
```

```python
train_conts = cont_grouped.groupby(["cycle", "state_full", "CAND_OFFICE_DISTRICT"])[["RepTotal", "RepCount", "DemTotal", "DemCount"]].sum().reset_index()
train_conts.head()
```

```python
super_competitive_df = pd.merge(super_competitive_df, train_conts, left_on=["year", "state", "district"], right_on=["cycle", "state_full", "CAND_OFFICE_DISTRICT"], how="left")
```

```python
train_supercomp = super_competitive_df[(super_competitive_df["year"] == 2020) | (super_competitive_df["year"] == 2018)][["year", "state", "district", "RepTotal", "RepCount", "DemTotal", "DemCount"]]
train_supercomp.fillna(0, inplace=True)
train_supercomp.head()
```

```python
train_supercomp["RepTotalShare"] = train_supercomp["RepTotal"] / (train_supercomp["RepTotal"] + train_supercomp["DemTotal"])
train_supercomp["RepCountShare"] = train_supercomp["RepCount"] / (train_supercomp["RepCount"] + train_supercomp["DemCount"])

train_supercomp.head()
```

```python
preds_2018_538 = pd.read_csv("delux_data/538_preds/house_district_forecast.csv")
preds_2018_538.head()
```

```python
final_deluxe_2018 = preds_2018_538[(preds_2018_538["model"] == "deluxe") & (preds_2018_538["forecastdate"] == "2018-11-06") & ((preds_2018_538["party"] == "R") | (preds_2018_538["party"] == "D"))]
final_deluxe_2018
```

```python
# Calculates Democratic and Republican vote shares for each district in 2018

final_deluxe_2018["RepVoteShare"] = (final_deluxe_2018["party"] == "R") * final_deluxe_2018["voteshare"]
final_deluxe_2018["DemVoteShare"] = (final_deluxe_2018["party"] == "D") * final_deluxe_2018["voteshare"]

del_grouped = final_deluxe_2018.groupby(["state", "district"])[["RepVoteShare", "DemVoteShare"]].sum().reset_index()

del_grouped["Rep2PVoteShare"] = del_grouped["RepVoteShare"] / (del_grouped["RepVoteShare"] + del_grouped["DemVoteShare"])

del_grouped["year"] = 2018

del_grouped.head()
```

```python
preds_2020_538 = pd.read_csv("delux_data/538_preds/house_district_toplines_2020.csv")
preds_2020_538.head()
```

```python
final_deluxe_2020 = preds_2020_538[(preds_2020_538["expression"] == "_deluxe") & (preds_2020_538["forecastdate"] == "11/3/20")]
final_deluxe_2020
```

```python
del_grouped_2020 = final_deluxe_2020[["district", "voteshare_mean_D1", "voteshare_mean_R1"]]
grp_2020 = del_grouped_2020["district"].str.split("-", expand=True).rename(columns={0: "state", 1: "district"})
grp_2020["Rep2PVoteShare"] = del_grouped_2020["voteshare_mean_R1"] / (del_grouped_2020["voteshare_mean_R1"] + del_grouped_2020["voteshare_mean_D1"])
grp_2020["year"] = 2020
grp_2020.head()
```

```python
preds_538 = pd.concat([del_grouped[["state", "district", "Rep2PVoteShare", "year"]], grp_2020])
```

```python
train_results = results_df.loc[(results_df["year"] == 2018) | (results_df["year"] == 2020),["year", "state", "district_num", "RepVotesMajorPercent"]]
train_results.head()
```

```python
train_results.loc[train_results["district_num"] == 0, "district_num"] = 1

train_results["district_num"] = train_results["district_num"].astype(int)
```

```python
import us

preds_538["state_full_name"] = preds_538["state"].apply(lambda x: us.states.lookup(x).name)

preds_538.district = preds_538.district.astype(int) 

preds_538.head()
```

```python
train_results = pd.merge(train_results, preds_538, left_on=["year", "state", "district_num"], right_on=["year", "state_full_name", "district"], how="left")

train_results.head()
```

```python
train_results = train_results[["year", "state_x", "district", "RepVotesMajorPercent", "Rep2PVoteShare"]].rename(columns={"state_x": "state"})

train_results.head()
```

```python
train_conts["super_comp"] = 1

train_conts["RepTotalShare"] = train_conts["RepTotal"] / (train_conts["RepTotal"] + train_conts["DemTotal"])
train_conts["RepCountShare"] = train_conts["RepCount"] / (train_conts["RepCount"] + train_conts["DemCount"])
```

```python
train_results = pd.merge(train_results, train_conts[["cycle", "state_full", "CAND_OFFICE_DISTRICT", "RepTotalShare", "RepCountShare"]], left_on=["year", "state", "district"], right_on=["cycle", "state_full", "CAND_OFFICE_DISTRICT"], how="left")

train_results.head()
```

```python
train_results["Rep2PVoteShare"] = train_results["Rep2PVoteShare"] * 100

train_results.head()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
models = [
    ("Linear Regression", LinearRegression()),
    ("Lasso", Lasso()),
    ("Ridge", Ridge()),
    ("Random Forest", RandomForestRegressor(n_estimators=100, max_depth=10)),
    ("Gradient Boosting", GradientBoostingRegressor(n_estimators=100, max_depth=10))
]
```

```python
model_infos = []

for model_stuff in models:

    model_info = {}

    model_info["name"] = model_stuff[0]

    model = model_stuff[1]

    model.fit(train_results[["Rep2PVoteShare"]], train_results["RepVotesMajorPercent"])
    model_info["model_base"] = model
    model_info["model_base_R2"] = model.score(train_results[["Rep2PVoteShare"]], train_results["RepVotesMajorPercent"])
    model_info["model_base_accuracy"] = ((model.predict(train_results[["Rep2PVoteShare"]]) > 50) == (train_results["RepVotesMajorPercent"] > 50)).mean()

    model_info["model_base_cv"] = cross_val_score(model, train_results[["Rep2PVoteShare"]], train_results["RepVotesMajorPercent"], cv=5)

    model.fit(train_results.loc[train_results["year"] == 2018,["Rep2PVoteShare"]], train_results.loc[train_results["year"] == 2018, "RepVotesMajorPercent"])
    model_info["model_base_2020_R2"] = model.score(train_results.loc[train_results["year"] == 2020,["Rep2PVoteShare"]], train_results.loc[train_results["year"] == 2020, "RepVotesMajorPercent"])
    model_info["model_base_2020_accuracy"] = ((model.predict(train_results.loc[train_results["year"] == 2020, ["Rep2PVoteShare"]]) > 50) == (train_results.loc[train_results["year"] == 2020, "RepVotesMajorPercent"] > 50)).mean()

    model.fit(train_comp_x, train_comp_y)
    model_info["model_full"] = model
    model_info["model_full_R2"] = model.score(train_comp_x, train_comp_y)
    model_info["model_full_accuracy"] = ((model.predict(train_comp_x) > 50) == (train_comp_y > 50)).mean()

    model_info["model_full_cv"] = cross_val_score(model, train_comp_x, train_comp_y, cv=5)

    model.fit(train_comp_x.loc[train_comp["year"] == 2018], train_comp_y.loc[train_comp["year"] == 2018])
    model_info["model_full_2020_R2"] = model.score(train_comp_x.loc[train_comp["year"] == 2020], train_comp_y.loc[train_comp["year"] == 2020])
    model_info["model_full_2020_accuracy"] = ((model.predict(train_comp_x.loc[train_comp["year"] == 2020]) > 50) == (train_comp_y.loc[train_comp["year"] == 2020] > 50)).mean()

    model_infos.append(model_info)


```
