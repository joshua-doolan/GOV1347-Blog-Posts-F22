```python
# This method gets the total FEC contributions from their API for a given cycle, list of committees, and list of zip codes

import requests
import json

def get_fec_contributions(cycle, committees, zips):
    url = f"https://api.open.fec.gov/v1/schedules/schedule_a/by_zip/?sort_null_only=false&sort_hide_null=false&sort_nulls_last=false&api_key=DEMO_KEY&sort=-total&per_page=100&cycle={cycle}&committee_id={committees}&zip={zips}"
    response = requests.get(url)
    return json.loads(response.text)
```

```python
zip_contribution_dfs = []
# i = 0
for ind, row in super_competitive_df[(super_competitive_df["year"] == 2018) | (super_competitive_df["year"] == 2020)].iterrows():
    # i += 1

    # if ind <= 451:
    #     continue

    race_zips = zcta_2018.loc[(zcta_2018["STATEFP"] == int(row["state_fips"])) & (zcta_2018["CD116FP"] == int(row["district"]))]["NAME20"]
    # print(int(row["district"]))
    # print(row["state_fips"])
    race_cands = cand_summ_df.loc[(cand_summ_df["CAND_OFFICE_ST"] == row["state_abb"]) & (cand_summ_df["CAND_OFFICE_DISTRICT"] == int(row["district"])) & (cand_summ_df["CYCLE"] == row["year"]) & (cand_summ_df["CAND_ID"].str.contains("H"))].sort_values("TTL_INDIV_CONTRIB", ascending=False)["CAND_ID"][:2]
    committee_ids = cand_link_df[(cand_link_df["CYCLE"] == row["year"]) & (cand_link_df["CAND_ID"].isin(race_cands))]["CMTE_ID"]


    api_url = "https://api.open.fec.gov/v1/schedules/schedule_a/by_zip/"

    if int(row["district"]) == 0:
        param_zips = None
    else:
        param_zips = list(race_zips)[:375]

    r = requests.get(
        api_url, 
        params={
            "api_key": os.environ.get("FEC_API_KEY"),
            "cycle": row["year"],
            "committee_id": list(committee_ids),
            "zip": param_zips,
            "sort_hide_null": False,
            "sort_null_only": False,
            "sort_nulls_last": False,
            "per_page": 100,
            "sort": "-total",
            "state": row["state_abb"]
        }
    )

    zip_contribution_df = pd.DataFrame(r.json()["results"])

    zip_contribution_dfs.append(zip_contribution_df)
    # print(race_zips[:3])
    # print(committee_ids)
```

```python
zip_contribution_df = pd.concat(zip_contribution_dfs)

zip_contribution_df.to_csv("zip_contributions_2.csv", index=False)
```
