```python
train_df = house_df[["State", "state_abb", "raceYear", "district_num", "RepVotesMajorPercent"]]
```

```python
train_df["national_turnout"] = 0
train_df.loc[train_df["raceYear"] == 2020, "national_turnout"] = 0.667
train_df.loc[train_df["raceYear"] == 2018, "national_turnout"] = 0.500
train_df.loc[train_df["raceYear"] == 2016, "national_turnout"] = 0.601
train_df.loc[train_df["raceYear"] == 2014, "national_turnout"] = 0.368
train_df.loc[train_df["raceYear"] == 2012, "national_turnout"] = 0.586
train_df.loc[train_df["raceYear"] == 2010, "national_turnout"] = 0.410
```

```python
train_df["dist_turnout"] = pd.merge(house_df, cvap_df, left_on=["State", "district_num", "raceYear"], right_on=["state", "cd", "year"], how="left")["TotalVotesMajor"]/pd.merge(house_df, cvap_df, left_on=["State", "district_num"], right_on=["state", "cd"], how="left")["cvap"]
```

```python
train_df["D_inc"] = house_df["DemStatus"] == "Incumbent"
train_df["R_inc"] = house_df["RepStatus"] == "Incumbent"
```

```python
train_df["avg_rating"] = pd.merge(train_df, expert_df, how="left", left_on=["raceYear", "State", "district_num"], right_on = ["year", "state", "district"])["avg_rating"]
```

```python
wiki_data = pd.read_html("https://en.wikipedia.org/wiki/2022_United_States_House_of_Representatives_elections")
```

```python
unique_districts = []
```

```python
for state_df in wiki_data[5:(5+50)]:
    for dist in state_df["District"]["Location"].unique():
        dist_data = {}
        dist_data["state"] = dist.split("\xa0")[0]
        dist_data["district"] = dist.split("\xa0")[1]
        dist_data["R_inc"] = ((state_df[state_df["District"]["Location"] == dist]["Incumbent"]["Party"] == "Republican") & (state_df[state_df["District"]["Location"] == dist]["Incumbent"]["Status"] == "Incumbent renominated")).sum() > 0
        dist_data["D_inc"] = ((state_df[state_df["District"]["Location"] == dist]["Incumbent"]["Party"] == "Democratic") & (state_df[state_df["District"]["Location"] == dist]["Incumbent"]["Status"] == "Incumbent renominated")).sum() > 0
        unique_districts.append(dist_data)
    # state = state_df.iloc[0, 0]
    # state_df.columns = state_df.iloc[0, :]
    # state_df = state_df.iloc[1:, :]
    # state_df["state"] = state
    # train_df = pd.concat([train_df, state_df], axis=0)
```

```python
test_df = pd.DataFrame(unique_districts)
test_df.loc[test_df["district"] == "at-large", "district"] = 0
test_df["district"] = test_df["district"].astype(int)
```
